```python
from yahooquery import Ticker
import pandas as pd

class AnalystRecommendationFetcher:
    def __init__(self, symbol):
        self.symbol = symbol
        self.ticker = Ticker(symbol)
        self.financial_data = {}

    def fetch_data(self):
        # Fetch financial data
        data = self.ticker.financial_data
        if isinstance(data, dict) and self.symbol in data:
            self.financial_data = data[self.symbol]
        else:
            self.financial_data = {}

    def get_data(self):
        if isinstance(self.financial_data, dict):
            return {
                'Symbol': self.symbol,
                'Current Price': self.financial_data.get('currentPrice'),
                'Target High Price': self.financial_data.get('targetHighPrice'),
                'Target Low Price': self.financial_data.get('targetLowPrice'),
                'Target Mean Price': self.financial_data.get('targetMeanPrice'),
                'Target Median Price': self.financial_data.get('targetMedianPrice'),
                'Recommendation Mean': self.financial_data.get('recommendationMean'),
                'Recommendation Key': self.financial_data.get('recommendationKey'),
                'Number of Analyst Opinions': self.financial_data.get('numberOfAnalystOpinions')
                # You can add more fields here as needed
            }
        else:
            return {
                'Symbol': self.symbol,
                'Error': 'Financial data not available or in unexpected format'
            }
```

```python
with open('companies.pkl', 'rb') as file:
    df_companies = pickle.load(file)

df_companies.head()
```

```python
stock_symbols = df_companies.Symbol  # Add any stocks you're interested in

# Fetch data for each stock
analyst_data_list = []
i = 0
for symbol in stock_symbols:
    i += 1
    analyst_fetcher = AnalystRecommendationFetcher(symbol)
    try:
        analyst_fetcher.fetch_data()
        analyst_data_list.append(analyst_fetcher.get_data())
    except:
        print(f"Couldn't fetch recommendations for stock {symbol}")
        
    print(f"Fetching data for symbol {i} out of {len(stock_symbols)}")

analyst_data_df = pd.DataFrame(analyst_data_list)
analyst_data_df
```

```python
with open('analystsRecommendations.pkl', 'wb') as file:
    pickle.dump(analyst_data_df, file)
```

```python
analyst_data_df.to_csv("analystsRecommendations.csv")
```
